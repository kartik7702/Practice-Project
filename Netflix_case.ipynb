# **DATA CLEANING AND ADJUSTING**

*   List item
*   List item




```python
 # treat time formate
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'])
netflix_df["date_added"] = pd.to_datetime(netflix_df['date_added'])
netflix_df['year_added'] = netflix_df['date_added'].dt.year
netflix_df['month_added'] = netflix_df['date_added'].dt.month
```

```python
netflix_df.head()
#date format is now adjested
```

# **Handling Missing Values**

rating,cast,country,



director have to be filled with "Unavalble"## Handling Missing Values

rating,cast,country,

director have to be filled with "Unavalble"

```python
netflix_df.fillna({'rating':"Unavailble",'cast':"Unavailble",'country':"Unavailble",'director':"Unavailble"}, inplace=True)
netflix_df.isna().sum()
```

```python
netflix_df[netflix_df.date_added.isnull()]
#CHECKING NULL VALUES IN DATE_ADDED
```

```python
resent_added_date=netflix_df['date_added'].max()
resent_added_date
#CHECKING THE RECENT DATE OF ADDING CONTENT
```

```python
netflix_df.fillna({'date_added':resent_added_date},inplace=True)
# REPLACING "NA" IN DATE ADDED WITH RECENT DATE
# AS WE HAVE OBSERVED THE MOST OF THE CONTENT WAS ADDED IN RESENT YEARS
```

```python
netflix_df.isna().sum()
```

no null value in "date_added"



```python
netflix_df[netflix_df.duration.isnull()]
#CHECKING null value in "duration" colomn
```

we can observe that all the null value in duration have something in common that is "director" named "Louis C.K."
Also we can see that missing data in duration column is availble in rating column.

```python
netflix_df.fillna({'duration':netflix_df.rating},inplace=True)
#filling na in duration column with data availble in rating column
```

```python
df_month = pd.DataFrame(netflix_df.month_added.value_counts()).reset_index().rename(columns={'index':'month','month_added':'count'})
# converting month number to month name
df_month['month_final'] = df_month['month'].replace({1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'June', 7:'July', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'})
df_month[:2]
```

```python
fig_month = px.funnel(df_month, x='count', y='month_final', title='Best month for releasing Content',
                      height=350, width=600, color_discrete_sequence=['#b20710'])
fig_month.update_xaxes(showgrid=False, ticksuffix=' ', showline=True)
fig_month.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))
fig_month.update_layout(margin=dict(t=60, b=20, l=70, r=40),
                        xaxis_title=' ', yaxis_title=" ",
                        plot_bgcolor='#333', paper_bgcolor='#333',
                        title_font=dict(size=25, color='#8a8d93', family="Lato, sans-serif"),
                        font=dict(color='#8a8d93'),
                        hoverlabel=dict(bgcolor="black", font_size=13, font_family="Lato, sans-serif"))
```
